```python
#Import Package
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Part 1: Basic Plotting Features
```

```python
#Draw Line Graph
xpoints = np.array([0, 6])#array is something like a vector/sequence
ypoints = np.array([0, 250])
plt.plot(xpoints, ypoints)
#the first parameters: points on X-axis, second: Y-axis. They should be in array form
plt.show()#display the graph
```

```python
#Pointsplot (Line Graph Without Line)
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.plot(xpoints, ypoints, "o", marker = "*")
#"o" indicates that there will be no line, marker specifies the points' shape
#For scatter plot we can include multiple points, but X-axis should have same number of points as Y-axis
#If don't specify X-axis' points, then as default will be 0, 1, 2, 3 etc., like indexes
plt.show()
```

```python
#Use Format to Change the Above Graph
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.plot(xpoints, ypoints, 'o:r', ms = 20, mec = 'b')
#'o:r' is the format. 
#format = marker\line\color, ms controls points' size, mec control points' edge's color, mfc controls points' fulfill color (not shown above)
plt.show()
```

```python
#Change Line Style
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, linestyle = 'dotted', color  = "r",linewidth = '20.5') 
#linestyle can be written as ls for short, dotted written as :, Ex: ls=":", controls the style of line
#color can be written as c for short, controls line's color
#linewidth can be written as lw for short, changes the width of the line
plt.show()
```

```python
#Stack Multiple Lines On One Graph
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 6])
y2 = np.array([6, 2, 7, 11])
plt.plot(x1, y1) 
plt.plot(x2, y2)
#Just add multiple plt.plot() functions
plt.show()
```

```python
#Creating Labels & Titles
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
#Create style to be applied later
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
#family: font, color: title&label color, size: font size
plt.title("Sports Watch Data", fontdict = font1, loc = 'left')
#Set label&title's format according to what has been specified
#loc: set title's position, only applicable for title
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.show()
```

```python
#Add Grid On Graph
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x, y)
#Add grid:
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
#Grid in green color, also set the grid line's style and width
#If add axis = "y", then grid only appears horizontal
plt.show()
```

```python
#Draw Subplots
#Plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)#plots will be arranged in 1 row, 2 cols, and this is the first one
plt.plot(x,y)
plt.title("SALES")#separate title for the first graph
#Plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")
#Add a whole title
plt.suptitle("MY SHOP")
plt.show()
```

```python
#Part 2: Different Kinds of Plots
```

```python
#1. Scatterplot
```

```python
#1.1: Single Scatterplot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
plt.show()
```

```python
#1.2: Compare Scatterplots
#Jusr add multiple scatter functions
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x1, y1)#color of points can be specified here by adding Ex.: color = "hotpink"
x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x2, y2)
#Different scatterplots' points will be in different colors
plt.show()
```

```python
#1.3: Scatterplot's Options
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
#Pass color and size as vectors to each matching points
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)#alpha controls the transparency of the points
plt.show()
```

```python
#2. Barplot
```

```python
#2.1: Vertical Barplot
x = np.array(["A", "B", "C", "D"])#categories
y = np.array([3, 8, 1, 10])#values of categories
plt.bar(x,y)
plt.show()
```

```python
#2.2: Horizontal Barplot
x = np.array(["A", "B", "C", "D"])#categories
y = np.array([3, 8, 1, 10])#values of categories
plt.barh(x,y)
plt.show()
```

```python
#2.3: Barplot's Options
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "hotpink", width = 0.1)#change the color and width of plots
#If using horizontal barplot, then use height instead of width
plt.show()
```

```python
#3. Histogram
```

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)#draw histogram of x
plt.show() 
```

```python
#4. Pie Chart
```

```python
#4.1: Single Pie Chart
y = np.array([350, 25, 25, 15])
#Pie chart will draw percentage of each values, starting from X-axis and rotate counterclockwise
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]#will serve as a vector, label matching points
plt.pie(y, labels = mylabels)
plt.show()
```

```python
#4.2: Pie Chart's Options
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
myexplode = [0.2, 0, 0, 0]#also serve as a vector, to explode each pies
plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
#startangle = 0: means starting from X-axis (again counterclockwise)
#shadow = True: add shadow to each pie
plt.legend(title = "Four Fruits:")#add legend with a title/header
plt.show()
```
